Giapetto Example from Operations Research Applications and Algorithms by W.L Winston



```python
import gurobipy as gp
from gurobipy import GRB
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = gp.Model('Giapetto')
```

```python
# Variables

x_1 = model.addVar(vtype=GRB.CONTINUOUS, name='x_1') # number of soldiers produced each week
x_2 = model.addVar(vtype=GRB.CONTINUOUS, name='x_2') # number of trains produced each week

# Objective Function

model.setObjective(3*x_1 + 2*x_2 , GRB.MAXIMIZE)

# Finishing Constraint
model.addConstr(2*x_1 + x_2 <= 100,'finishing')
# Carpentary Constraint
model.addConstr(x_1 + x_2 <= 80,'carpentary')
# Demand for soldiers
model.addConstr(x_1 <= 40,'soldier_demand')
```

```python
model.optimize()
```

```python
# Results

print('Obj: %g' % model.objVal)
print('x_1 = %g' % x_1.X)
print('x_2 = %g' % x_2.X)
```
